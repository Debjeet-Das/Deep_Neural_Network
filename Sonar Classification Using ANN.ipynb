```python
import pandas as pd
from keras.layers import Dense, Flatten, Dropout
from keras.models import Sequential
```

```python
data = pd.read_csv("sonar.csv")
data.head()
```

```python
# Creating X and Y variables
X = data.iloc[:,:-1].values
Y = data.iloc[:,-1].values

Y_hot = pd.get_dummies(Y )
```

```python
# Splitting the data 
from sklearn.model_selection import train_test_split
trainx,testx,trainy,testy = train_test_split(X,Y_hot,test_size = 0.2,)
```

```python
# initialize 
classifier = Sequential()

# Adding layer
classifier.add(Dense(units = 200,input_dim=60,activation = "relu",kernel_initializer = "he_uniform"))
#classifier.add(Dropout(0.2))

classifier.add(Dense(units = 100,activation = "relu"))
#classifier.add(Dropout(0.02))

classifier.add(Dense(units = 50,activation = "relu"))
#classifier.add(Dropout(0.08))

classifier.add(Dense(units = 20,activation = "relu"))
#classifier.add(Dropout(0.02))

classifier.add(Dense(units = 2,activation = "softmax"))

# Compiling

classifier.compile(loss="categorical_crossentropy",optimizer = "adam",metrics =["accuracy"])


# Fit
classifier.fit(trainx,trainy,epochs=100)
```

```python
# Predicting 
pred = classifier.predict(testx)
pred = pred.round()
pred
```

```python
# Accuracy of the model
from sklearn.metrics import accuracy_score
accuracy_score(pred,testy)
```

```python
# Evaluating the model
classifier.evaluate(testx,testy)
```
