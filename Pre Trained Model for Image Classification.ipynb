```python
#Already trained model
from keras.models import model_from_json

#later

#load json and create model

json_file=open("CatDogModel.json","r")
loaded_model_json=json_file.read()
json_file.close()

# Loading the model
loaded_model=model_from_json(loaded_model_json)
loaded_model.load_weights("CatDogWeights.h5")



# Test image

import numpy as np
from keras.preprocessing import image
test_image = image.load_img('C:\\Users\\Debjeet Das\\Desktop\\PYTHON\\P16-Deep-Learning-AZ\\Deep_Learning_A_Z\\Volume 1 - Supervised Deep Learning\\Part 2 - Convolutional Neural Networks (CNN)\\dataset\\single_prediction\\cat_or_dog_1.jpg', target_size = (64, 64))
test_img = image.img_to_array(test_image)
test_img = np.expand_dims(test_image, axis = 0)


result=loaded_model.predict(test_img)


if result[0][0] == 1:
    prediction = 'dog'
else:
    prediction = 'cat'

print("It Is a ",prediction)


import matplotlib.pyplot as plt
plt.imshow(test_image)
```
